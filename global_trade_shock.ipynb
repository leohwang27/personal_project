```python
import pandas as pd

df_raw = pd.read_csv("TradeData_11_24_2025_13_5_30.csv", encoding="latin1")
df_raw.columns = df_raw.columns.str.replace('ï»¿', '')
df_raw.head()
```

```python
df = df_raw.copy()

# 1) Keep only exports
df = df[df["flowCode"] == "X"]          # 'X' usually = Export

# 2) Keep only partner = USA
df = df[df["partnerDesc"] == "USA"]     # or "United States of America" if that’s how it shows

# 3) Keep only the reporters you care about (optional if you already filtered in Comtrade)
target_reporters = ["China", "Rep. of Korea", "Japan", "Viet Nam", "India"]
df = df[df["reporterDesc"].isin(target_reporters)]

df.head()
```
